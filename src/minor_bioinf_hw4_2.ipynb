```python
import pandas as pd

c1 = pd.read_csv('SRR3414635.counts', sep="\t",  names=['geneID', 'c1']).astype(str)
c2 = pd.read_csv('SRR3414636.counts', sep="\t",  names=['geneID', 'c2']).astype(str)
c3 = pd.read_csv('SRR3414637.counts', sep="\t",  names=['geneID', 'c3']).astype(str)
r1 = pd.read_csv('SRR3414629.counts', sep="\t",  names=['geneID', 'r1']).astype(str)
r2 = pd.read_csv('SRR3414630.counts', sep="\t",  names=['geneID', 'r2']).astype(str)
r3 = pd.read_csv('SRR3414631.counts', sep="\t",  names=['geneID', 'r3']).astype(str)

c1 = c1.merge(c2, how='left', on='geneID')
c1 = c1.merge(c3, how='left', on='geneID')
c1 = c1.merge(r1, how='left', on='geneID')
c1 = c1.merge(r2, how='left', on='geneID')
c1 = c1.merge(r3, how='left', on='geneID')

ALL_cnts = c1[:-5]
ALL_cnts
```

```python
ALL_cnts.to_csv('ALL.counts', sep='\t', index=False)
```
